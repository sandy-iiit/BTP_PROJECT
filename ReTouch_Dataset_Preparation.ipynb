Delete Files and Folders


```python
for dataset in ['Cirrus','Spectralis','Topcon']:

    root_directory = './RETOUCH-TrainingSet-'+dataset

    for root, dirs, files in os.walk(root_directory, topdown=False):
        # Delete files with .npy extension
        npy_files = [f for f in files if f.endswith('.npy')]
        for file in npy_files:
            file_path = os.path.join(root, file)
            os.remove(file_path)
            print(f"Deleted file: {file_path}")

        # Delete 'resized' folder
        resized_folder = os.path.join(root, 'resized')
        if os.path.exists(resized_folder):
            shutil.rmtree(resized_folder)
            print(f"Deleted folder: {resized_folder}")

        # Delete files with .tif extension
        npy_files = [f for f in files if f.endswith('.tif')]
        for file in npy_files:
            file_path = os.path.join(root, file)
            os.remove(file_path)
            print(f"Deleted file: {file_path}")

        # Delete 'oct' folder
        resized_folder = os.path.join(root, 'oct')
        if os.path.exists(resized_folder):
            shutil.rmtree(resized_folder)
            print(f"Deleted folder: {resized_folder}")

        # Delete 'reference' folder
        resized_folder = os.path.join(root, 'reference')
        if os.path.exists(resized_folder):
            shutil.rmtree(resized_folder)
            print(f"Deleted folder: {resized_folder}")

        # Delete folders ending with "png" along with their contents
        png_folders = [d for d in dirs if d.endswith('png')]
        for folder in png_folders:
            folder_path = os.path.join(root, folder)
            shutil.rmtree(folder_path)
            print(f"Deleted folder: {folder_path}")
```

# Spectralis


## Spectralis .raw to .npy conversion


```python
def convert(img, target_type_min, target_type_max, target_type):
    imin = img.min()
    imax = img.max()

    a = (target_type_max - target_type_min) / (imax - imin)
    b = target_type_max - a * imax
    new_img = (a * img + b).astype(target_type)
    return new_img
```

```python
dataset = 'Spectralis'
for i in range(25,49):
    file_num = str(i)
    if len(file_num) < 2:
        file_num += '0'
        file_num = file_num[::-1]

    input_file = './RETOUCH-TrainingSet-'+dataset+'/TRAIN0'+file_num+'/oct.raw'
    npimg1= np.fromfile(input_file, dtype=np.uint16)
    npimg=convert(npimg1, 0, 255, np.uint8)
    print(file_num+'. '+'oct.raw-shape-------->'+str(np.shape(npimg)))
    npy_data = np.zeros((496,512,49),dtype=np.uint8)
    reshape_train_data = np.resize(npimg, (49,496,512))
    for i in range(0,49):
        npy_data[:,:,i] = reshape_train_data[i,:,:]# ----------------------Try again without this step to see result--------->
    print(file_num+'. '+'oct.npy-shape-------->'+str(np.shape(npy_data)))
    np.save('./RETOUCH-TrainingSet-'+dataset+'/TRAIN0'+file_num+'/oct.npy', npy_data)

    input_file = './RETOUCH-TrainingSet-'+dataset+'/TRAIN0'+file_num+'/reference.raw'
    npimg = np.fromfile(input_file, dtype=np.uint8)
    print(file_num+'. '+'reference.raw-shape-->'+str(np.shape(npimg)))
    npy_ref = np.zeros((496,512,49),dtype=np.uint8)
    reshape_train_data = np.resize(npimg, (49,496,512))
    for i in range(0,49):
        npy_ref[:,:,i]=reshape_train_data[i,:,:]
    print(file_num+'. '+'reference.npy-shape-->'+str(np.shape(npy_ref)))
    np.save('./RETOUCH-TrainingSet-'+dataset+'/TRAIN0'+file_num+'/reference.npy', npy_ref)
```

```python
for dataset in ['Cirrus','Spectralis','Topcon']:

    root_directory = './RETOUCH-TrainingSet-'+dataset

    for root, dirs, files in os.walk(root_directory, topdown=False):
        # Delete files with .npy extension
        npy_files = [f for f in files if f.endswith('.npy')]
        for file in npy_files:
            file_path = os.path.join(root, file)
            os.remove(file_path)
            print(f"Deleted file: {file_path}")

        # Delete 'resized' folder
        resized_folder = os.path.join(root, 'resized')
        if os.path.exists(resized_folder):
            shutil.rmtree(resized_folder)
            print(f"Deleted folder: {resized_folder}")

        # Delete files with .tif extension
        npy_files = [f for f in files if f.endswith('.tif')]
        for file in npy_files:
            file_path = os.path.join(root, file)
            os.remove(file_path)
            print(f"Deleted file: {file_path}")

        # Delete 'oct' folder
        resized_folder = os.path.join(root, 'oct')
        if os.path.exists(resized_folder):
            shutil.rmtree(resized_folder)
            print(f"Deleted folder: {resized_folder}")

        # Delete 'reference' folder
        resized_folder = os.path.join(root, 'reference')
        if os.path.exists(resized_folder):
            shutil.rmtree(resized_folder)
            print(f"Deleted folder: {resized_folder}")

        # Delete folders ending with "png" along with their contents
        png_folders = [d for d in dirs if d.endswith('png')]
        for folder in png_folders:
            folder_path = os.path.join(root, folder)
            shutil.rmtree(folder_path)
            print(f"Deleted folder: {folder_path}")
```

```python
destination_directory = '../Images'
os.mkdir(destination_directory)
os.mkdir(destination_directory+'/OCT')
os.mkdir(destination_directory+'/Reference')

for i in range(1,71):
    if i < 25:
        dataset = 'Cirrus'
    elif i < 49:
        dataset = 'Spectralis'
    else:
        dataset = 'Topcon'

    file_num = str(i)
    if len(file_num) < 2:
        file_num += '0'
        file_num = file_num[::-1]

    path = './RETOUCH-TrainingSet-'+dataset+'/TRAIN0'+file_num
    oct_path = path+'/oct'
    reference_path = path+'/reference'

    for file in sorted(os.listdir(oct_path)):
        source_path = oct_path+'/' + file
        destination_path = destination_directory+'/OCT/'+ file_num + '_' + file[-7:]
        shutil.copy(source_path, destination_path)
        print(file)

    for file in sorted(os.listdir(reference_path)):
        source_path = reference_path+'/' + file
        destination_path = destination_directory+'/Reference/'+ file_num + '_' + file[-7:]
        shutil.copy(source_path, destination_path)
        print(file)
```
